```python
import os
from hcqm import make_specs
from model import predict_global_bpm, load_model
import time
correct = 0
total = 0

audio_dir = 'data/giantsteps-tempo-dataset/audio/'
bpm_dir = 'data/giantsteps-tempo-dataset/annotations_v2/tempo/'

audio_files = os.listdir(audio_dir)
model = load_model('deeprhythm0.1.pth', device='cpu')
specs = make_specs(device='cpu')
times = []
for file in audio_files:
    audio_path = audio_dir+file
    bpm_path = bpm_dir+file.replace('.mp3', '.bpm')
    with open(bpm_path, 'r') as bpm_file:
        true_bpm = float(bpm_file.read().strip())
    start = time.time()
    pred_bpm = predict_global_bpm(audio_path, model=model, specs=specs)
    delta_t = time.time()-start
    times.append(delta_t)
    print(f'pred: {pred_bpm} actual: {true_bpm}')
    total += 1
    lower_bounds = true_bpm * 0.96
    upper_bounds = true_bpm * 1.04

    # Update correct predictions
    is_correct = (pred_bpm >= lower_bounds) & (pred_bpm <= upper_bounds)
    if is_correct:
        correct += 1
accuracy = 100 * correct / total
print(f'Acc1 on giantsteps: {accuracy:.2f}%')
average_time = sum(times) / len(times)
min_time = min(times)
max_time = max(times)
print(f'Average prediction time: {average_time:.4f}s')
print(f'Minimum prediction time: {min_time:.4f}s')
print(f'Maximum prediction time: {max_time:.4f}s')
```
